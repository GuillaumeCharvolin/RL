## Agent Random

Take each time a random machine according to an uniform distribution.

```python
new_agent_uniform = AgentUniform("AgentUniform", new_space_normal, trials_number=1000)

new_agent_uniform.play()

new_agent_uniform.visualize_gain_variability()
```

```python
new_agent_uniform.calculate_optimal_strategy_gain()
new_agent_uniform.visualize_gain(show_optimal=True)

print(new_agent_uniform.gain_history)
print(new_agent_uniform.optimal_gain_history)
print(new_agent_uniform.space.best_machine())
```

## Agent Training

Take a training percentage parameter, training_percent of the trials are doing randomly according to an uniform distribution, then exploit the machine that gave the best gain.

```python
### training_percent = 0.5

new_agent_training = AgentTraining("AgentTraining0.5", new_space_normal, trials_number=1000, training_percent=0.5)

new_agent_training.play()

new_agent_training.visualize_gain_variability()

new_agent_training.calculate_optimal_strategy_gain()

print(new_space_normal)
print(new_agent_training.space_information)

new_agent_training.visualize_gain(show_optimal=True)
```

```python
### training_percent = 0.25

new_agent_training = AgentTraining("AgentTraining0.25", new_space_normal, trials_number=1000, training_percent=0.25)
new_agent_training.play()
new_agent_training.visualize_gain_variability()
new_agent_training.calculate_optimal_strategy_gain()
new_agent_training.visualize_gain(show_optimal=True)
```

## Epsilon greedy approach

Always try the machine that has the best results from now with an epsilon chance to try something else.

```python
### Epsilon = 0.1

new_agent_epsilon = AgentEpsilonGreedy("AgentEpsilonGreedy0.1", new_space_normal, trials_number=1000, epsilon=0.1)

new_agent_epsilon.play()

new_agent_epsilon.visualize_gain_variability()

new_agent_epsilon.calculate_optimal_strategy_gain()
new_agent_epsilon.visualize_gain(show_optimal=True)
```

```python
### Epsilon = 0.01

new_agent_epsilon = AgentEpsilonGreedy("AgentEpsilonGreedy0.01", new_space_normal, trials_number=1000, epsilon=0.01)

new_agent_epsilon.play()

new_agent_epsilon.visualize_gain_variability()

new_agent_epsilon.calculate_optimal_strategy_gain()
new_agent_epsilon.visualize_gain(show_optimal=True)
```

```python
### Epsilon = 0.2

new_agent_epsilon = AgentEpsilonGreedy("AgentEpsilonGreedy0.2", new_space_normal, trials_number=1000, epsilon=0.2)

new_agent_epsilon.play()

new_agent_epsilon.visualize_gain_variability()

new_agent_epsilon.calculate_optimal_strategy_gain()
new_agent_epsilon.visualize_gain(show_optimal=True)
```

```python
### Epsilon = 0.2

new_agent_epsilon = AgentEpsilonGreedy("AgentEpsilonGreedy0.2D0.99", new_space_normal, trials_number=1000, epsilon=0.2, decaying=0.99)

new_agent_epsilon.play()

new_agent_epsilon.visualize_gain_variability()

new_agent_epsilon.calculate_optimal_strategy_gain()
new_agent_epsilon.visualize_gain(show_optimal=True)
```
